```python
import pandas as pd

file_path1 = "Schreibmaschinen.csv"
file_path2 = "Liste_AK Kommunikation.csv"

```

```python
# Erstmal mit Latin1 (ISO-8859-1), oft Standard bei Windows/Excel
schreibmaschienen_df = pd.read_csv(file_path1, encoding="cp1252", sep=";",header=None, skiprows=1)
kommunikation_df = pd.read_csv(file_path2, encoding="cp1252", sep=";",header=None, skiprows=1)

```

```python
# Überblick
print(schreibmaschienen_df.shape)
print(schreibmaschienen_df.columns)
schreibmaschienen_df.head()
```

```python
# Überblick
print(kommunikation_df.shape)
print(kommunikation_df.columns)
kommunikation_df.head()
```

```python
# Nur die ersten 300 Zeilen behalten
kommunikation_df300 = kommunikation_df.head(300)
kommunikation_df300.info
```

```python

# Untereinander anhängen
df_combined = pd.concat([kommunikation_df300,schreibmaschienen_df], ignore_index=True)

print(df_combined)
```

```python
# 2. Werte in Spalte 4 bereinigen (/ -> -)
df_combined.iloc[:, 4] = df_combined.iloc[:, 4].astype(str).str.replace("/", "-", regex=False)

# 3. Nach Spalte 4 eindeutige Zeilen behalten
df_unique = df_combined.drop_duplicates(subset=df_combined.columns[4]).reset_index(drop=True)

print(df_unique.shape)
print(df_unique.head())
```

```python
df_unique.to_csv("fertig.csv", index=False, encoding="utf-8")
```
